```python
#Group IV-VI semiconductors

data1 = mpr.query(criteria = 
                 {
                    'elements': 
                         {'$in': ['Ge','Sn','Pb'], '$all': ['S'], 
                         } ,'nelements': 2, #'anonymous_formula': {'A':1,'B':1},
                 },
                 properties = ['pretty_formula','band_gap'])

data2 = mpr.query(criteria = 
                 {
                     'elements': 
                         {'$in': ['Ge','Sn','Pb'], '$all': ['Se'],
                         } ,'nelements': 2, #'anonymous_formula': {'A':1,'B':1},
                 },
                 properties = ['pretty_formula','band_gap'])

data3 = mpr.query(criteria = 
                 {
                     'elements': 
                         {'$in': ['Ge','Sn','Pb'], '$all': ['Te'],
                         } ,'nelements': 2, #'anonymous_formula': {'A':1,'B':1},
                 },
                 properties = ['pretty_formula','band_gap'])
```

```python
data.extend(data1)

data.extend(data2)

data.extend(data3)

data
```

```python
features = []

for i in range(0, len(data)):
    comp =  pym.Composition(data[i]['pretty_formula'])
    
    #Calculating the number of each element
    elements_in_comp = comp.to_data_dict['elements']
    number_of_elements = comp.to_data_dict['unit_cell_composition']
    
    #Calculating the electronegativity difference
    e_neg_diff = abs(number_of_elements[elements_in_comp[0]]*pym.Element(elements_in_comp[0]).X\
                           - number_of_elements[elements_in_comp[1]]*pym.Element(elements_in_comp[1]).X)
    
    #Atomic fraction calculation
    atom_fract = comp.get_atomic_fraction(elements_in_comp[0])
    
    #Calculating the molecular weight
    molec_weight = comp.weight
    
    #Obtaining the group number of the cation
    group_num = pym.Element(elements_in_comp[0]).group
    
    features.append([comp, e_neg_diff, atom_fract, molec_weight, group_num])
```

```python
df_1 = pd.DataFrame.from_dict(data)
```

```python
df_2 = pd.DataFrame(features, columns = ['comp','eneg_diff','atom_frac','molec_weight','group_num'])
```

```python
df = pd.concat([df_1, df_2.reindex(df_1.index)], axis=1)
```
